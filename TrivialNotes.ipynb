```python
import networkx as nx
from networkx.algorithms import bipartite

# Construct the bipartite weighted graph projection of nodes A,B,C,D,E,F 
# and find the weight of the edge (A,C).
# What is the weight of the edge (A,C)?

B = nx.Graph() # creating graph
B.add_edges_from([('A', 'G'),('A','I'), ('B','H'), ('C', 'G'), ('C', 'I'),('D', 'H'), ('E', 'I'), ('F', 'G'), ('F', 'J')])
X1 = set(['A', 'B', 'C', 'D', 'E', 'F']) # putting all nodes into a set


# projection of bipartite graph onto set 
# of nodes X1 based on main graph B
# Projection is done to find common nodes btw 2 graph
P = bipartite.weighted_projected_graph(B,X1)

print(P.edges['A','C']) # find weight of the edge AC
print(P.edges(data = True)) # list all edges with weights, wo the data = true, just list all edges

# P.edges['A']['C'] # get dict of attributes of edge AC
# P.edges['A']['C']['weight'] # get specific attribute's weight
# P.edges['C']['A'] # if undirected order doesnt matter


# G.edges['A']['B'] # will output nested dict of 
# {0 : {'relation': 'family', 'weight': 6} , 
#  1 : {'relation': 'friend', 'weight': 15}}
# IF undirected, order does not matter remeber, indexing is as such ['A']['B'][0]['weight'] will return   6
G = nx.MultiGraph()
G.add_edge('A','B', weight = 6, relation = 'family')
G.add_edge('A','B', weight = 15, relation = 'friend')
G.add_edge('C','B', weight = 7, relation = 'friend')
G.nodes(data = True)
```
